# Busienss Transformation and Modeling
 

```python
query = """
SELECT 
  ROW_NUMBER() OVER(ORDER BY pi.start_date, pi.product_number) AS product_key,
  pi.product_id,
  pi.product_number,
  pi.product_name,
  pi.category_id,
  pc.category,
  pc.subcategory,
  pc.maintenance_flag,
  pi.product_cost,
  pi.product_line,
  pi.start_date
FROM silver.crm_prd_info pi
LEFT JOIN silver.erp_px_cat_g1v2 pc
  ON pi.category_id = pc.category_id
WHERE pi.end_date IS NULL
"""

df = spark.sql(query)
```

```python
df.limit(10).display()
```

# Write to Gold Table


```python
df.write.mode("overwrite").format("delta").saveAsTable("gold.dim_products")
```
