# Busienss Transformation and Modeling

```python
query = """
SELECT 
    sd.order_number, 
    p.product_key,
    c.customer_key,
    sd.order_date,
    sd.ship_date,
    sd.due_date,
    sd.sales_amount,
    sd.quantity,
    sd.price
FROM silver.crm_sales_details sd
LEFT JOIN gold.dim_products p
    ON sd.product_number = p.product_number
LEFT JOIN gold.dim_customers c
    ON sd.customer_id = c.customer_id
"""

df = spark.sql(query)
```

```python
df.display()
```

# Write to Gold Table

```python
df.printSchema()
```

```python
df.write.mode("overwrite").format("delta").saveAsTable("gold.fact_sales")
```
